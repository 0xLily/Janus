```python
# Install GEOparse if not installed
# pip install GEOparse

import GEOparse

# --- 1. Download GSE112672 ---
gse_id = "GSE112672"
print(f"Downloading {gse_id} ...")

gse = GEOparse.get_GEO(
    gse_id,
    destdir=".",          # save files to current directory
    annotate_gpl=False
)

print("Download completed.\n")

# --- 2. Print basic metadata of the GSE ---
print("===== GSE Metadata =====")
print("Title:", gse.metadata.get("title"))
print("Summary:", gse.metadata.get("summary"))
print("Overall design:", gse.metadata.get("overall_design"))
print("\n")

# --- 3. List GSM samples and their characteristics ---
print("===== GSM Samples & Characteristics =====")
for gsm_name, gsm in gse.gsms.items():
    print(f"Sample: {gsm_name}")
    for item in gsm.metadata.get("characteristics_ch1", []):
        print("  -", item)
    print("\n")
```

```python
import GEOparse
import textwrap

gse = GEOparse.get_GEO("GSE112672", destdir=".", annotate_gpl=False)

# Wrap GSE summary
summary = gse.metadata.get("summary", [""])[0]
print("===== Summary =====")
print(textwrap.fill(summary, width=80))
print("\n")

# Wrap GSM sample characteristics
print("===== GSM Samples & Characteristics =====")
for gsm_name, gsm in gse.gsms.items():
    print(f"Sample: {gsm_name}")
    for item in gsm.metadata.get("characteristics_ch1", []):
        print(textwrap.fill("  - " + item, width=80))
    print()
```
